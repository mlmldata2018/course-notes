```python
%matplotlib notebook
import numpy as np
from scipy import stats
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.figure()
plt.plot(df['OXYGEN'],df['CTDPRS'],'.')
plt.gca().invert_yaxis()

plt.xlabel('$O_2 [\mu M]$')
plt.ylabel('pressure[dbar]')
```

```python
x = df['NITRAT']
y = df['PHSPHT']

results = stats.linregress(x,y)
```

```python
results
```

```python
plt.figure()
plt.plot(df['NITRAT'],df['PHSPHT'],'o')
```

```python
ii=(np.isfinite(x) & np.isfinite(y))
results=stats.linregress(x[ii],y[ii])
results
```

```python
plt.figure()
plt.plot(df['NITRAT'][ii],df['PHSPHT'][ii],'o')
plt.plot(np.array([0,45]),results.slope*np.array([0,45])+results.intercept,'r-')
```

```python
plt.figure()
plt.plot(df['PH_TOT'],df['CTDPRS'], 'o')
plt.gca().invert_yaxis()

plt.xlabel('pH')
plt.ylabel('pressure[dbar]')
```
